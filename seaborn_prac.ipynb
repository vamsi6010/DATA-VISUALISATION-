```python
tips = sns.load_dataset('tips')
```

```python
tips.dropna()
```

```python
sns.relplot(x='total_bill',y='tip',data = tips)
#relational plot
#from plot we can observe the relation between tips and total bill
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, hue= 'smoker', style= 'time')
#hue differentiates rhe data based on color 
#style differetiates the data based on shape
#from plot we can say that person who paid the highest tip was a smoker and he visited for dinner
```

```python
sns.relplot(x='total_bill',y='tip',style='size',data=tips)
#size is the size of the group came to eat
```

```python
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```python
fmri = sns.load_dataset('fmri')
fmri
```
